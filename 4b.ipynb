```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv("C://Users/Meenakshi/Downloads/ML/Mall_Customers - Mall_Customers.csv")
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Genre']=le.fit_transform(df['Genre'])
df.head(10)
```

```python
from sklearn.cluster import KMeans
wcss=[]
for k in range(2,11):
    kmeans=KMeans(n_clusters=k,random_state=42)
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)
```

```python
import seaborn as sns
sns.lineplot(x=range(2,11),y=wcss)
```

```python
from scipy.cluster.hierarchy import dendrogram,linkage
from sklearn.cluster import AgglomerativeClustering
```

```python
matrix=linkage(df,method="ward")
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,7))
dendrogram(matrix,orientation="top",show_leaf_counts=True)
plt.show()
```

```python
agglo=AgglomerativeClustering(n_clusters=4,linkage="ward")
df['cluster']=agglo.fit_predict(df)
```

```python
sns.scatterplot(data=df,x=df['Annual Income (k$)'],y=df['Spending Score (1-100)'],hue="cluster')
```
